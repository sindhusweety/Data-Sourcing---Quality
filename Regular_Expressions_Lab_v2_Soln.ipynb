### Exclusion Pattern Matching
Author a RE that can provide a match for the terms in column 1 while
 at the same time not match the terms in column 2.  Prove your solution
 works by iterating over the lists and assessing the success or failure
 of the pattern.
```
Expected Outcome:
```

Develop a **single RE** that will match:

All of these | None of these
--------------------|-------------------
afoot | Atlas
catfoot | Aymoro
dogfoot | Iberic
fanfoot | Mahran
foody | Ormazd
foolery | Silipan

````python
import re

match_str='''\
afoot,Atlas
catfoot,Aymoro
dogfoot,Iberic
fanfoot,Mahran
foody,Ormazd
foolery,Silipan
foolish,altared
fooster,chandoo
footage,crenel
foothot,crooked
footle,fardo
footpad,folksy
footway,forest
hotfoot,hebamic
jawfoot,idgah
mafoo,manlike
nonfood,marly
padfoot,palazzi
prefool,sixfold
sfoot,tarrock
unfool,unfold'''

re_pattern = r'foo'

# compile for efficiency
rc = re.compile(re_pattern)

for pair in match_str.split('\n'):
    
    should, should_not = pair.split(',')
    
    # if match obj is returned, there was a match. Else, not so much...
    m1 = rc.search(should)
    m1 = 'and does' if m1 is not None else 'but does not'

    m2 = rc.search(should_not)
    m2 = 'and does not' if m2 is None else 'but does'

    print(f'RE: {re_pattern}. {should} should match {m1}.')
    print(f'RE: {re_pattern}. {should_not} should not match {m2}.')
````

### Regular Expressions - Finding Email Addresses
You work for SPAM Marketing Co.  Your supervisor has just received a
 text file (VancouverDataTextMiningData.txt) containing, what she
 believes, will be pure SPAM marketing gold - previously un-spammed
 email addresses.  I realize that email RE patterns are widely
 available.  For the purposes of skill building, please author your
 own, original, email pattern.

You have been asked to author a Python script to extract all of the
 email addresses from that text file.

### Logical Requirements
1. open the file for reading
2. for each line in the file, use a RE to scan the line for valid
 email address.  According the addr-spec standard as defined in RFC
 5322, the general format of an email address is local-part@domain.top-
 level domain. The part before the @ symbol (local-part) identifies the
 name of a mailbox. The part after the @ symbol consists of a domain
 and top-level domain. For this lab you only need to find valid email
 addresses.

**NOTE:** The 'VancouverDataTextMiningData.txt' text file contains some
 pretty messy data; however, it is a UTF-8 encoded file.
 Open the file and iterate over the lines using your RE to search for valid email addresses.

### Deliverable
Print the list of matched email addresses (one email per line).

````python
import re

# TEST DATA
in_file = 'VancouverDataTextMiningData.txt'

# REs written assuming re.IGNORECASE | re.VERBOSE
# RE Explained:
# \b - begin at word boundaries (matches at a position that is followed by a word character
#      but not preceded by a word character, or that is preceded by a word character but not
#      followed by a word character.)
# 1st character class is the pattern for the local-part of an email.
# + is 1 or more of these chars
# @ matches literal @
# 2nd char class is the pattern for the domain of an email
# + is 1 or more of these chars
# \. matches literal . (to match 'dot' it must be protected as . is a metachar. could use [.]
# 3d char class is the pattern for the top-level domain of an email
# quantifier {2,} matches 2 or more such characters
# \b - end at word boundaries


# This pattern will match 99.9% of all emails - I'm guessing...
e_valid_str = r'\b[A-Z0-9._%+-]+@[A-Z0-9.-]+\.[A-Z]{2,}\b'

# slup in all emails
# open using context manager
with  open(in_file, mode='r', encoding='utf-8') as fh:

    # file handles are iterable
    for line in fh:
        # strip off trailing new-line chars
        line = line.rstrip('\n')

        # YOUR CODE HERE.
        # line is a string that may (or may not) contain one or more email addresses

        # Alternative 1
        # findall() returns a list of all matching text (not a match object)
        for email in re.findall(e_valid_str, line, re.IGNORECASE):
            print(f'Found email(1): {email}')

        # Alternative 2
        # finditer() returns an iterator over all matching match objects
        for email_match in re.finditer(e_valid_str, line, re.IGNORECASE):
            print(f'Found email(2): {email_match.group()}')
````

## Regular Expressions - Pig Latin Too
### THIS IS FOR EXPOSITION PURPOSES ONLY!
The moral here - choose your weapon wisely.  RegExs (exclusively) are not a good tool choice
to solve this problem.

This function receives a (word/string) as a parameter. The objective
 of the function is to convert the incoming string to its "Pig Latin"
 equivalent and return the Pig Latin string.  Yes, this is the same
 exercise used in the Python strings lab; however, this time you are
 required to use RE methods to complete the text manipulations.

For our purposes, we will use the following as the rules for
 translation of a word into "Pig Latin":

1. A word is a consecutive sequence of letters (a-z, A-Z) or
 apostrophes. You may assume that the input to the function will only
 be a single "word". Examples: Zebra , doesn't , apple
2. If a word starts with a vowel, the Pig Latin version is the
 original word with "way" added to the end
3. If a word starts with a consonant, or a series of consecutive
 consonants, the Pig Latin version transfers ALL consonants up to the
 first vowel to the end of the word, and adds "ay" to the end.
4. The letter 'y' should be treated as a consonant if it is the
 first letter of a word, but treated as a vowel otherwise.
5. If the original word is capitalized, the new Pig Latin version of
 the word should be capitalized in the first letter.  If the original
 capital letter was a consonant, and thus moved, it should not be
 capitalized once in its new location.

### Logical Requirements
1. Author your 'ToPigLatin' function
2. Convert the following test data into a Python dictionary.

Test Word (keys) | Pig Latin (values)
-----------------|-------------------
football | ootballfay
Pittsburgh | Ittsburghpay
Apple | Appleway
oink | oinkway
ontology | ontologyway
yellow | ellowyay
yttrium | iumyttray

3. Iterate over the test data and invoke your *ToPigLatin* function.
4. Confirm your returned results against the values in the second column

### Deliverable
For each test word, print the test word and its Pig Latin translation.
Print each pair on a separate line.

````python
# Pig Latin Converter Code
import re

VOWELS = 'aeiou'

##################################
# NOTE:
# Part of the point of this problem is demonstrate that one should
# choose the 'right tool for job' mentality when it comes to data wrangling.
#
# REs are NOT the right tool for this particular job.
# The RE solution below is easily twice as difficult to implement as it's
# text sequence counterpart.
##################################

# TEST DATA
test_dict = dict(
    football = 'ootballfay',
    Pittsburgh = 'Ittsburghpay',
    Apple = 'Appleway',
    oink = 'oinkway',
    ontology = 'ontologyway',
    yellow = 'ellowyay',
    yttrium = 'iumyttray')

# test boundary condition.  word that starts with
# y but is followed by a consonant
# yttrium - a silvery, malleable metallic element that is found in the same ores as
# other rare-earth elements

def to_pig_latin(eng_word):
    if len(eng_word) == 0:
        raise ValueError("Input must contain characters.")

    # remember if I need to return title case word
    m = re.match(r'^[A-Z]', eng_word)
    if m:
        to_upper = True
        first_char = m[0]
    else:
        m = re.match(r'^[a-z]', eng_word, re.IGNORECASE)
        if m:
            to_upper = False
            first_char = m[0]
        else:
            raise ValueError("Input must start with a character.")

    # now, normalize the word to LC
    eng_word = re.sub('[A-Z]', lambda m: m.group().lower(), eng_word)

    # low hanging fruit. capture letters after 1st char
    m = re.match(r'^[aeiou](.*)$', eng_word, re.IGNORECASE)
    if m:
        rest = m.group(1)
        return f'{first_char}{rest}way'

    # ok, word starts with consonant or 'y'

    cons_str = ''
    # find all characters one at time
    c_list = re.findall(r'.', eng_word)

    for i,c in enumerate(c_list):
        if i == 0 and c == 'y':
            cons_str += 'y'
        else:
            # matching a single char here
            m = re.search(r'^[aeiouy]', c)
            # not a vowel
            if m is None :
                cons_str += c
            else:
                break
    # This else: is associated with the for:
    # In short, the else for a for: is entered if the for loop
    # terminates normally.  That is, the loop is NOT exited because of a 'break'
    # Yes, 'else' is less than ideal syntax for this feature
    else:
        raise ValueError(f'Very few English words do not contain a vowel:{eng_word}')

    # Note: rf is a raw f-string
    re_pat = rf'^{cons_str}(.*)$'
    m = re.match(re_pat, eng_word, re.IGNORECASE)

    # rearrange the word.  group(1) is what currently follows cons_str
    new_word = m.group(1) + cons_str + 'ay'

    # The 'x if condition else y' is known as a conditional statement
    # Other languages (c, java, etc.) refer to this type of statement as a ternary statement
    return new_word.title() if to_upper else new_word

def main():
    # ingest test data
    # Split the multiline string at the newlines
    for k,v in test_dict.items():
        pig = to_pig_latin(k)
        if pig == v:
            print(f'{k} => {pig}')
        else:
            print(f'Failure for Enlish word: {k}. {pig} should be {v}.')

# IMO, the if below is the most "hackish" thing in python.  In a nutshell, .py files
# can either be standalone scripts or the could be modules to be imported.
# How can you tell if a .py script is being imported or if it is being invoked as a
# standalone script? By checking the __name__ property.
# If python has set the __name__ to "__main__"  you know the script has not been
# imported but rather ran as a script.
# See https://www.freecodecamp.org/news/whats-in-a-python-s-name-506262fe61e8/

# In a JNB, a cell is "ran", not imported, which is why this works here
if __name__ == "__main__":
    main()
````
